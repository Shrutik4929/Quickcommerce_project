```python
import pandas as pd
```

```python
df = pd.read_csv("/content/quick_commerce_data_raw.csv")
```

```python
df.head()
```

```python
df.info()
df.describe()
```

```python
print("Duplicate Rows:", df.duplicated().sum())
```

```python
df.isnull().sum()
```

```python
df['Items_Count'] = df['Items_Count'].fillna(df['Items_Count'].median())
```

```python
df['Items_Count'] = df['Items_Count'].astype(int)
```

```python
df['Customer_Rating'] = df['Customer_Rating'].fillna(df['Customer_Rating'].median())
```

```python
df['Discount_Applied'] = df['Discount_Applied'].fillna(0)
df['Discount_Applied'] = df['Discount_Applied'].astype(int)
```

```python
df['Delivery_Partner_Rating'] = df['Delivery_Partner_Rating'].fillna(df['Delivery_Partner_Rating'].median())
```

```python
df.isnull().sum()
```

```python

df.dropna(subset=['City'], inplace=True)

print(df.isnull().sum())
```

```python
df.dtypes
```

```python
df['Order_Value'] = df['Order_Value'].round(2)
```

```python
df['Order_Value']
```

```python
df['Delivery_Time_Min']
```

```python
df['Delivery_Time_Min'] = df['Delivery_Time_Min'].round()
```

```python
df['Delivery_Time_Min'] = df['Delivery_Time_Min'].astype(int)
```

```python
df['Delivery_Time_Min']
```

```python
df['Customer_Rating'] = df['Customer_Rating'].round(2)
```

```python
df['Age_Group'] = pd.cut(df['Customer_Age'],
    bins=[17, 25, 35, 50, 100],
    labels=['Young Adult', 'Adult', 'Middle Age', 'Senior'])
```
